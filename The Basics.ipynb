```julia
#ChatGPT Answer
using Distributions, Plots

# Define the Binomial distribution
bd = Binomial(10, 0.5)

# Compute and print the PMF for each k
for k in 0:10
    println("k = ", k, " p = ", pdf(bd, k))
end

# Define the Beta distribution and a range of p values
betad = Beta(2, 5)  # Example Beta distribution
p = range(0, 1, length=100)  # Define a range for p

# Compute the PDF values for Beta distribution
betapdf = pdf.(betad, p)  # Use broadcasting with `.` to apply pdf element-wise

# Plot the Beta distribution PDF
plot(p, betapdf, label="Beta(2,5) PDF", xlabel="p", ylabel="Density")
```
